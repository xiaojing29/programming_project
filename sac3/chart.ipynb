```python
import matplotlib.pyplot as plt
import numpy as np

# Data
models = [
    "Gemma-7B\n(7B Parameters)",
    "Llama-3-8B\n(8B Parameters)",
    "Mistral-7B\n(7B Parameters)",
    "Phi-3-mini\n(3.8B Parameters)",
    "Qwen1.5-7B\n(7B Parameters)"
]
self_check_scores = {
    "Gemma-7B": [50, 50, 50],
    "Llama-3-8B": [60, 70, 50],
    "Mistral-7B": [48, 66, 60],
    "Phi-3-mini": [80, 80, None],
    "Qwen1.5-7B": [44, 48, 56]
}
cross_check_scores = {
    "Gemma-7B": [56, 56],
    "Llama-3-8B": [67, 76],
    "Mistral-7B": [94, 94, 68],
    "Phi-3-mini": [86, 90],
    "Qwen1.5-7B": [64, 80]
}

# Setting indices and labels for different test configurations
self_check_labels = ["Self-Check (3 Responses)", "Self-Check (5 Responses)", "Self-Check (10 Responses)"]
cross_check_labels = ["Cross-Check (5 Perturbation Questions)", "Cross-Check (10 Perturbation Questions)"]

# Prepare bar indices
n_models = len(models)
bar_width = 0.15
x_self = np.arange(n_models)
x_cross = x_self + bar_width * 3

fig, ax = plt.subplots(figsize=(12, 8))

# Plot bars for self-check consistency
for i, label in enumerate(self_check_labels):
    y_values = [self_check_scores[model.split("\n")[0]][i] if self_check_scores[model.split("\n")[0]][i] is not None else 0 for model in models]
    ax.bar(x_self + i * bar_width, y_values, bar_width, label=label)

# Plot bars for cross-check consistency
for i, label in enumerate(cross_check_labels):
    y_values = [cross_check_scores[model.split("\n")[0]][i] if cross_check_scores[model.split("\n")[0]][i] is not None else 0 for model in models]
    ax.bar(x_cross + i * bar_width, y_values, bar_width, label=label)

# Set axis labels and chart title
ax.set_xlabel("Models")
ax.set_ylabel("AUROC (%)")
ax.set_title("AUROC Scores for Self-Check and Cross-Check Consistency")
ax.set_xticks(x_self + bar_width * 1.5)
ax.set_xticklabels(models)
ax.legend()

plt.xticks(rotation=20)
plt.tight_layout()
plt.savefig('model_performance_comparison_with_params.png')
plt.show()
```
