```python
import matplotlib.pyplot as plt

# Data
models = ["Gemma-7B", "Llama-3-8B", "Mistral-7B", "Phi-3-mini", "Qwen1.5-7B"]
parameters = [7, 8, 7, 3.8, 7]
self_check = [70, 72, 60, 80, 65]
cross_check = [80, 85, 74, 88, 78]

# Plot
fig, ax = plt.subplots(figsize=(10, 6))
bar_width = 0.35
index = range(len(models))

bar1 = ax.bar(index, self_check, bar_width, label='Self-Check AUROC')
bar2 = ax.bar([i + bar_width for i in index], cross_check, bar_width, label='Cross-Check AUROC')

ax.set_xlabel('Models')
ax.set_ylabel('AUROC (%)')
ax.set_title('Model Performance Comparison')
ax.set_xticks([i + bar_width / 2 for i in index])
ax.set_xticklabels(models)
ax.legend()

plt.savefig('/mnt/data/model_performance.png')
plt.show()
```
